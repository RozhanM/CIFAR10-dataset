importing dataset

```python
from tensorflow.keras.datasets import cifar10
```

```python
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

images = np.zeros((1000, 32, 32, 3))
labels = np.zeros((1000,))


for i in range(10):
    class_num = np.where(y_train == i)[0]
    np.random.shuffle(class_num)
    selected_num = class_num[:100]
    images[i*100:(i+1)*100] = x_train[selected_num]
    labels[i*100:(i+1)*100] = i


```

```python
x_train.shape
```

```python
x_train = x_train.reshape(x_train.shape[0], -1) / 255.
```

```python
x_train.shape
```

making W and B

```python
w = np.random.randn(3072 ,10)
```

```python
scores = np.dot(x_train, w)
```

```python
b = np.random.randn(10,1)
```

```python
s = (scores.T + b).T
```

```python
s.shape
```

calculating hinge loss

```python
margin = s - y_train  + 1
Li = np.maximum(0, margin)
```

```python
Li
```

```python
L = Li.sum() / 5000
L
```
